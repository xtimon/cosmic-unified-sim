````python
# Stage 4: N-body Gravitational Dynamics
print("üåç STAGE 4: GRAVITATIONAL DYNAMICS")
print("-"*60)

# Create a mini solar system to demonstrate gravity
G = 6.67430e-11
M_sun = 1.989e30
AU = 1.496e11

# Sun and Earth
sun = Body(name="Sun", mass=M_sun, 
           position=np.array([0.0, 0.0, 0.0]),
           velocity=np.array([0.0, 0.0, 0.0]))

# Earth at 1 AU with circular velocity
v_earth = np.sqrt(G * M_sun / AU)
earth = Body(name="Earth", mass=5.97e24,
             position=np.array([AU, 0.0, 0.0]),
             velocity=np.array([0.0, v_earth, 0.0]))

# Simulate for 1 year
sim = NBodySimulator([sun, earth])
year_seconds = 365.25 * 24 * 3600
times, states = sim.simulate(t_span=(0, year_seconds), n_points=365)

# Check energy conservation
E0, dE = sim.get_energy_conservation()
print(f"   System: Sun + Earth")
print(f"   Simulation: 1 year")
print(f"   Energy conservation: {abs(dE)*100:.8f}% change")
print()

# Calculate Earth's orbital period from simulation
r = np.sqrt(states[3,:]**2 + states[4,:]**2 + states[5,:]**2)
print(f"   Earth orbit: {r.min()/AU:.4f} - {r.max()/AU:.4f} AU")
print(f"   ‚úÖ Stable orbit achieved!")
print()
print(f"   üåü From matter ‚Üí stars ‚Üí planets ‚Üí life!")

# Store for later
orbital_data = {'times': times, 'states': states}
````

## üåç Prediction 1: Solar System Stability (1 Million Years)

## üåå Prediction 3: Milky Way - Andromeda Collision

In ~4.5 billion years, our galaxy will collide with Andromeda (M31). Using N-body simulation to model this cosmic event!

````python
# Prediction 3: Milky Way - Andromeda Collision
print("üåå PREDICTION 3: MILKY WAY - ANDROMEDA COLLISION")
print("="*70)
print()

# Galaxy parameters (simplified to supermassive black holes + mass centers)
# Real collision involves billions of stars, but we model gravitationally bound cores

# Scale: 1 unit = 1 kpc, time in Myr
# Masses in solar masses, using gravitational parameter

# Milky Way: ~1.5 √ó 10^12 M_sun (total including dark matter)
# Andromeda: ~1.2 √ó 10^12 M_sun (total)
# Distance: ~2.5 million light years = ~765 kpc
# Approach velocity: ~110 km/s

# Simplified N-body: Galaxy cores + satellite particles
M_MW = 1.5e12  # Solar masses
M_M31 = 1.2e12  # Solar masses
M_sun = 1.989e30  # kg
G = 6.67430e-11  # m^3/(kg s^2)

# Convert units: kpc, Myr
kpc = 3.086e19  # meters
Myr = 1e6 * 365.25 * 24 * 3600  # seconds

# G in kpc^3 / (M_sun * Myr^2)
G_scaled = G * M_sun * Myr**2 / kpc**3  # ~4.5e-3

print("INITIAL CONDITIONS:")
print(f"  Milky Way mass:    {M_MW:.1e} M_sun")
print(f"  Andromeda mass:    {M_M31:.1e} M_sun")
print(f"  Current distance:  765 kpc (2.5 million light-years)")
print(f"  Approach velocity: ~110 km/s")
print()

# Create galaxy cores
mw_core = Body(name='Milky_Way', mass=M_MW * M_sun,
               position=np.array([0.0, 0.0, 0.0]) * kpc,
               velocity=np.array([0.0, 0.0, 0.0]) * 1000)  # m/s

# Andromeda approaching
m31_core = Body(name='Andromeda', mass=M_M31 * M_sun,
                position=np.array([765.0, 0.0, 0.0]) * kpc,
                velocity=np.array([-110.0, 50.0, 0.0]) * 1000)  # m/s (with transverse component)

# Add some "satellite" particles around each galaxy (simplified star clusters)
np.random.seed(42)
n_satellites = 8

satellites = []
for i in range(n_satellites // 2):
    # MW satellites
    angle = np.random.uniform(0, 2*np.pi)
    r = np.random.uniform(30, 80)  # kpc
    v_orbital = np.sqrt(G * M_MW * M_sun / (r * kpc)) / 1000  # km/s
    
    sat_mw = Body(name=f'MW_sat_{i}', mass=1e9 * M_sun,  # Small mass
                  position=np.array([r * np.cos(angle), r * np.sin(angle), 0.0]) * kpc,
                  velocity=np.array([-v_orbital * np.sin(angle), v_orbital * np.cos(angle), 0.0]) * 1000)
    satellites.append(sat_mw)
    
    # M31 satellites
    angle = np.random.uniform(0, 2*np.pi)
    r = np.random.uniform(30, 80)  # kpc
    v_orbital = np.sqrt(G * M_M31 * M_sun / (r * kpc)) / 1000  # km/s
    
    sat_m31 = Body(name=f'M31_sat_{i}', mass=1e9 * M_sun,
                   position=np.array([765.0 + r * np.cos(angle), r * np.sin(angle), 0.0]) * kpc,
                   velocity=np.array([-110 - v_orbital * np.sin(angle), 50 + v_orbital * np.cos(angle), 0.0]) * 1000)
    satellites.append(sat_m31)

# Create simulation
all_bodies = [mw_core, m31_core] + satellites
galaxy_sim = NBodySimulator(all_bodies)

# Simulate 10 billion years
t_collision = 10e9 * 365.25 * 24 * 3600  # 10 Gyr in seconds
print(f"Simulating {10} billion years of galactic dynamics...")
times_gal, states_gal = galaxy_sim.simulate(t_span=(0, t_collision), n_points=500)

print("Done!")
print()

# Analyze collision
times_Gyr = times_gal / (1e9 * 365.25 * 24 * 3600)
mw_pos = states_gal[0:3, :] / kpc
m31_pos = states_gal[3:6, :] / kpc

distance = np.sqrt((mw_pos[0] - m31_pos[0])**2 + 
                   (mw_pos[1] - m31_pos[1])**2 +
                   (mw_pos[2] - m31_pos[2])**2)

# Find first close approach
min_idx = np.argmin(distance)
first_approach_time = times_Gyr[min_idx]
min_distance = distance[min_idx]

print("COLLISION TIMELINE:")
print(f"  First close approach: {first_approach_time:.2f} billion years from now")
print(f"  Minimum separation:   {min_distance:.1f} kpc")
print(f"  Final separation:     {distance[-1]:.1f} kpc")
print()
print("FATE OF THE GALAXIES:")
if distance[-1] < 100:
    print("  -> Galaxies MERGE into 'Milkomeda'!")
else:
    print("  -> Galaxies in extended interaction/merger process")
print()
print("NOTE: Earth will likely survive! Space between stars is vast.")
````
